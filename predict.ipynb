```python
X = train.loc[:,['Open']]
x_train, x_cv, y_train, y_cv = train_test_split(X,train.High)

#training the model
clf = LinearRegression()
clf.fit(x_train,y_train)

#predicting on cv

pred = clf.predict(x_cv)

#calculating mse

mse = py.mean((pred - y_cv)**2)

r_sq = clf.score(x_cv,y_cv)

print("R-square ",r_sq)

## calculating coefficients

coeff = DataFrame(x_train.columns)

coeff['Coefficient Estimate'] = Series(clf.coef_)
print(coeff)
```

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, clf.predict(x_train), color = 'blue')
plt.title('Open vs High (Training set)')
plt.xlabel('open')
plt.ylabel('High')
plt.show()
```

```python
plt.scatter(x_cv, y_cv, color = 'red')
plt.plot(x_cv, clf.predict(x_cv), color = 'blue')
plt.title('Open vs High (Testing set)')
plt.xlabel('Open')
plt.ylabel('High')
plt.show()
```
